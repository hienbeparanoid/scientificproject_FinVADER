```python

import pandas as pd
import os
```

```python
def merge_files(directory, keyword):
    # Get the current working directory
    cwd = os.getcwd()
    # Create the path to the 'cmt' directory
    directory_path = os.path.join(cwd, directory)

    # Get a list of all the file paths in the directory
    file_paths = [os.path.join(directory_path, filename) for filename in os.listdir(directory_path) if filename.endswith('.csv') and keyword in filename]

    # Read each file into a DataFrame and add it to the list
    dfs = [pd.read_csv(file_path) for file_path in file_paths]

    # Concatenate all the DataFrames in the list
    df = pd.concat(dfs, ignore_index=True)

    return df
```

```python
aapl_merged = merge_files('cmt', 'aapl')
aapl_merged.head()
```

```python
amd_merged = merge_files('cmt', 'amd')
amd_merged.head()
```

```python
msft_merged = merge_files('cmt', 'msft')
msft_merged.head()
```

```python
nio_merged = merge_files('cmt', 'nio')
nio_merged.head()
```

```python
tsla_merged = merge_files('cmt', 'tsla')
tsla_merged.head()
```

```python
aapl_merged.to_csv('aapl_merged.csv', index=False)
amd_merged.to_csv('amd_merged.csv', index=False)
msft_merged.to_csv('msft_merged.csv', index=False)
nio_merged.to_csv('nio_merged.csv', index=False)
tsla_merged.to_csv('tsla_merged.csv', index=False)
```
