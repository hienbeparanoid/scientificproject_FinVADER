```python
# 2 list chứa index là tín hiệu sell/buy của df data_tsla, lấy ở kq bên scapestockdata.py
final_buy = [130, 4, 135, 12, 17, 147, 22, 157, 29, 36, 42, 47, 175, 51, 188, 64, 71, 80, 213, 93, 227, 107, 111, 119]
final_sell = [0, 132, 5, 140, 16, 18, 152, 26, 30, 38, 166, 44, 49, 180, 60, 67, 196, 73, 203, 85, 220, 99, 108, 116, 122]
```

```python
stockk = pd.read_csv('data_tsla.csv')
# Initialize the 'signal' column with 0
stockk['signal'] = 0

# Set 'signal' to 1 for the indices in 'final_buy'
stockk.loc[stockk.index.isin(final_buy), 'signal'] = 1

# Set 'signal' to 0 for the indices in 'final_sell'
stockk.loc[stockk.index.isin(final_sell), 'signal'] = -1

stockk.to_csv('data_tsla_signal.csv')
stockk
```

```python
class finVADERnSMA1(Strategy):
    def init(self):
        pass

    def next(self):
        signal = self.data.signal[-1]
        if signal == 1:
            self.position.close()
            self.buy()
        elif signal == -1:
            self.position.close()
            self.sell()
```

```python
stock = pd.read_csv('data_tsla_signal.csv', index_col='Datetime', parse_dates=True)
#stock = stock.rename(columns={'Datetime': 'date'})

# Sử dụng Backtest với dữ liệu mới
bt2 = Backtest(stock, finVADERnSMA1, cash=100000, commission=.002)
stat2 = bt2.run()
bt2.plot()
stat2
```
